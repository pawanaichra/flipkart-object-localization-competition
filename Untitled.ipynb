```python
%matplotlib tk
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def init():  # only required for blitting to give a clean slate.
    line.set_ydata([np.nan] * len(x))
    return line,


def animate(i):
    line.set_ydata(np.sin(x + i / 100))  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, init_func=init, interval=1, blit=True, save_count=50)

plt.show()
```

```python
import cv2
import torch
import numpy as np
image = cv2.imread("images\\2.png")
edges = cv2.Canny(image,100,200)
kernel = np.ones((5,5), np.uint8)
img_dilation = cv2.dilate(edges, kernel, iterations=1)
img_erosion = cv2.erode(img_dilation, kernel, iterations=1)
img1 = cv2.medianBlur(img_erosion,3)
img_dilation = cv2.dilate(img1, kernel, iterations=1)
img2=img_dilation.reshape(1,640, 480)
image = image.reshape(3,640,480)
image = np.concatenate((image, img2), axis = 0)
image = torch.from_numpy(image).double()
img=image.type(torch.DoubleTensor).double().cuda()
img
```
